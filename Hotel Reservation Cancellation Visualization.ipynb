# Data Analysis & Visualization

```python
cancelled_perc=df['is_canceled'].value_counts(normalize=True)
print(cancelled_perc)

plt.figure(figsize= (3,2))
plt.title('Reservation Status Count')
plt.bar(['NotCancelled', 'Canceled'], df['is_canceled'].value_counts(), edgecolor='k', width=0.7)
plt.show()
```

```python
# Ensure 'is_canceled' is treated as categorical with appropriate labels

df['is_canceled'] = df['is_canceled'].astype(str)
plt.figure(figsize=(8, 4))
ax1 = sns.countplot(x='hotel', hue='is_canceled', data=df, palette='Blues')

# Set the legend title and location
ax1.legend( bbox_to_anchor=(1, 1))

plt.title('Reservation Status in Different Hotels', size=20)
plt.ylabel('Number of Reservations')
plt.show()
```

```python
# Filter rows where hotel is 'Resort Hotel'
resort_hotel = df[df['hotel'] == 'Resort Hotel']

# Check value counts of 'is_canceled' in the filtered DataFrame
resort_hotel['is_canceled'].value_counts(normalize=True)
```

```python
# Filter rows where hotel is 'City Hotel'
city_hotel = df[df['hotel'] == 'City Hotel']

# Check value counts of 'is_canceled' in the filtered DataFrame
city_hotel['is_canceled'].value_counts(normalize=True)
```

```python
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize=(20,8))
plt.title('Average Daily Rate in City & Resort Hotel', fontsize = 30)
plt.plot(resort_hotel.index, resort_hotel['adr'], label = 'Resort Hotel')
plt.plot(city_hotel.index, city_hotel['adr'], label = 'City Hotel')
plt.legend(fontsize = 20)
plt.show()
```

```python
df['month'] = df['reservation_status_date'].dt.month
plt.figure(figsize=(16, 8))
ax1 = sns.countplot(x='month', hue='is_canceled', data=df, palette='bright')
legend_labels, _ = ax1.get_legend_handles_labels()  # Corrected method name
ax1.legend(bbox_to_anchor=(1, 1))
plt.title('Reservation Status Per Month', size=20)
plt.xlabel('Month')
plt.ylabel('Number of Reservations')
plt.legend(['Not Canceled', 'Canceled'])
plt.show()
```

```python
plt.figure(figsize= (12,7))
plt.title('ADR Per Month for Canceled Reservations', fontsize=20)
sns.barplot(x='month', y='adr', data=df[df['is_canceled'] == '1'].groupby('month')['adr'].sum().reset_index(), palette='bright')
plt.xlabel('Month')
plt.ylabel('ADR')
plt.show()
```

```python
cancelled_data= df[df['is_canceled'] == '1']
top_10_countries = cancelled_data['country'].value_counts()[:10]
plt.figure( figsize = (8,8))
plt.title('Top 10 Countries with Reservation Cancelled')
plt.pie(top_10_countries, autopct = '%.2f', labels= top_10_countries.index)
plt.show()
```

```python
# Calculate mean ADR for cancelled reservations
cancelled_df_adr = cancelled_data.groupby('reservation_status_date')['adr'].mean().reset_index()
cancelled_df_adr.sort_values('reservation_status_date', inplace=True)

# Filter non-cancelled reservations
not_cancelled_data = df[df['is_canceled'] == '0']
not_cancelled_df_adr = not_cancelled_data.groupby('reservation_status_date')['adr'].mean().reset_index()
not_cancelled_df_adr.sort_values('reservation_status_date', inplace=True)

# Plotting
plt.figure(figsize=(20, 6))
plt.title('Average Daily Rate', size=20)
plt.plot(not_cancelled_df_adr['reservation_status_date'], not_cancelled_df_adr['adr'], label='Not Cancelled') 
plt.plot(cancelled_df_adr['reservation_status_date'], cancelled_df_adr['adr'], label='Cancelled') 
plt.legend()
plt.show()
```

```python

# For cancelled reservations
cancelled_data = df[df['is_canceled'] == '1']
cancelled_df_adr = cancelled_data.groupby('reservation_status_date')[['adr']].mean().reset_index()

# For not cancelled reservations
not_cancelled_data = df[df['is_canceled'] == '0']
not_cancelled_df_adr = not_cancelled_data.groupby('reservation_status_date')[['adr']].mean().reset_index()

# Filter the date range
filtered_cancelled_df_adr = cancelled_df_adr[(cancelled_df_adr['reservation_status_date'] > '2016-01-01') & 
                                             (cancelled_df_adr['reservation_status_date'] < '2017-09-01')]

filtered_not_cancelled_df_adr = not_cancelled_df_adr[(not_cancelled_df_adr['reservation_status_date'] > '2016-01-01') & 
                                                     (not_cancelled_df_adr['reservation_status_date'] < '2017-09-01')]

# Step 3: Plot Data
plt.figure(figsize=(20, 6))
plt.title('Average Daily Rate', size=20)

# Plotting not cancelled reservations
plt.plot(filtered_not_cancelled_df_adr['reservation_status_date'], filtered_not_cancelled_df_adr['adr'], label='Not Cancelled')

# Plotting cancelled reservations
plt.plot(filtered_cancelled_df_adr['reservation_status_date'], filtered_cancelled_df_adr['adr'], label='Cancelled')

plt.legend(fontsize=20)
plt.xlabel('Date')
plt.ylabel('Average Daily Rate')
plt.show()
```
